```python
def is_contained(low1: int, hi1: int, low2: int, hi2: int, debug=False):
    """
    Returns true if list is contained within other list

    list 1 within list 2:
    2-3, 2-4

    list 2 within list 1:
    2-4, 2-3

    """

    low1 = int(low1)
    low2 = int(low2)
    hi1 = int(hi1)
    hi2 = int(hi2)

    if debug:
        print(f"{low1} >= {low2}: {low1 >= low2}")
        print(f"{hi1} <= {hi2}: {hi1 <= hi2}")
        print(f"{low2} >= {low1}: {low2 >= low1}")
        print(f"{hi2} <= {hi1}: {hi2 <= hi1}")

    return ((low1 >= low2) and (hi1 <= hi2)) or ((low2 >= low1) and (hi2 <= hi1))


is_contained(2,3,2,4, True)
```

```python
def overlap(l1, h1, l2, h2):
    """
    Return true if ranges overlap at all

    e.g. 
        5-7, 7-9
        2-8, 3-7


    """
    l1 = int(l1)
    l2 = int(l2)
    h1 = int(h1)
    h2 = int(h2)

    return not (h1 < l2 or l1 > h2)

overlap(5, 7, 7, 9)
```
