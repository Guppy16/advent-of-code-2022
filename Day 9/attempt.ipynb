```python
def move_h(move, h):
    """
    Move: U,D,R,L
    h: pos
    """

    # Move the head
    if move == "U":
        h[1] += 1
    elif move == "D":
        h[1] -= 1
    elif move == "L":
        h[0] -= 1
    elif move == "R":
        h[0] += 1
    else:
        raise RuntimeError("Move not known: ", move)

h = [0,0]
move_h("U", h)
print(h)
```

```python
def move_t(h, t):
    """
    h: pos
    t: pos
    """

    # Check if T is above H
    # if h == t:
    #     return
    
    # Check if T is with one of H
    if abs(h[0] - t[0]) <= 1 and abs(h[1] - t[1]) <= 1:
        return

    # Else T is at least 2 cells away form H

    """
    x # # # x
    # - - - #
    # - H - #
    # - - - #
    x # # # x
    """

    # IF T LEFT of H, move 1 right
    if t[0] < h[0]:
        t[0] += 1

    # if T RIGHT of H
    if t[0] > h[0]:
        t[0] -= 1

    # if T ABOVE H
    if t[1] > h[1]:
        t[1] -= 1
    
    # if T BELOW H
    if t[1] < h[1]:
        t[1] += 1
```

```python
h_pos = [0,0]   # x,y
t_pos = [0,0]   # x,y

visited_pos = set()

with open("test_data.txt") as f:
    for line in f:

        # line -> move, count
        move, count = line.split()

        for _ in range(int(count)):
            # Move h
            move_h(move, h_pos)

            # Move t
            move_t(h_pos, t_pos)

            # Keep track of positions moved by tail
            visited_pos.add(tuple(t_pos))

            print(h_pos, t_pos)
        
        # break

print(len(visited_pos))
```

```python
# Part 2

# Simulate with 10 knots
n = 10
x_pos = [[0,0] for _ in range(n)]

# NOTE: x_pos[0] is HEAD; x_pos[-1] is TAIL

visited_pos = set()

with open("test_data.txt") as f:
    for line in f:

        # line -> move, count
        move, count = line.split()

        for _ in range(int(count)):
            # Move head
            move_h(move, x_pos[0])

            for h_pos, t_pos in zip(x_pos, x_pos[1:]):
                # Move t
                move_t(h_pos, t_pos)

            print(t_pos)
            # Keep track of positions moved by tail
            visited_pos.add(tuple(t_pos))

print(len(visited_pos))
```
