```python
def get_filestore_from_cwd(filestore: dict, cwd: list):
    """
    return the filestore from the cwd
    """
    fstore = filestore
    for folder in cwd:
        fstore = fstore[folder]
    
    return fstore
    
# Test this
filestore = {"/" : {"a" : {"b" : {"c" : 1}}} }
cwd = ["/", "a", "b"]

print(get_filestore_from_cwd(filestore, cwd))
```

```python
def parse_text(line, filestore, cwd):
    """
    Parse text

    Assumptions
    - if a directory has been navigated to, then it contains all subdirectories
    
    """
    # if cd
    if "$ cd" in line:
        _, _, folder = line.split()

        if folder == "..":
            # Move up one directory
            cwd.pop()
        
        elif folder == "/":
            # Move to head
            cwd = [cwd[0]]
        
        else:
            cwd.append(folder)

        # filestore[folder] = {}
        # return filestore[folder]

    # if ls
    elif "$ ls" in line:
        return
        return filestore

    # previous command was ls
    else:
        begin, end = line.split()
        fstore = get_filestore_from_cwd(filestore, cwd)

        # if dir
        if begin == "dir":
            fstore[end] = {}
        
        # else it is a file
        else:
            fstore[end] = int(begin)

        # return filestore
```

```python
# Traverse filestore

# folder limit
folder_limit = 100000

total = 0

# Perform a DFS to find a file
def get_filestore_sum(filestore: dict=filestore):
    global total

    # keep track of folder size
    folder_size = 0

    # Get key value pair
    for k,v in filestore.items():
        
        # if v is an int, it is a filesize
        if type(v) == int:
            folder_size += v
            # return v

        # if v is a dict, it is a folder
        # we want to get the total folder size, 
        # and determine if it it less than max_size
        elif type(v) == dict:
            sub_folder_size = get_filestore_sum(v)
            if (sub_folder_size < folder_limit):
                total += sub_folder_size
            print(sub_folder_size)

            folder_size += sub_folder_size

        else:
            raise RuntimeError(k,v)
    
    if (folder_size < folder_limit):
        total += folder_size
    print(folder_size)
    return folder_size
    
    # Return condition

total_size = get_filestore_sum(filestore)
print(total_size)
```
